```python
import re

def is_english(text):
    english_pattern = re.compile(r'^[A-Za-z0-9.,!?;:(){}\[\]<>@#$%^&*_+=\-|\\/~`"\'£€¥₹₽\s]*$')
    return bool(re.match(english_pattern, text))
```

```python
def remove_after_hyphen(text):
    last_hyphen_index = text.rfind('-')  # Find the index of the last hyphen

    if last_hyphen_index != -1:  # If hyphen is found
        extracted_text = text[:last_hyphen_index]  # Extract text before the last hyphen
    else:
        extracted_text = text  # If no hyphen is found, keep the original text
    return extracted_text
```

```python
def numhyphnumb(query):
    return(re.sub(r'\d-\d.*?\s', '', query))
```

```python
# remove digits from a senetence
import re
def remove_numbers(query):
    clean_text = re.sub(r'\d+', '', query)
    # print(clean_text)
    return clean_text
```

```python
for index, data in df['TranslatedIngredients'].iteritems():
    # print(data)
    data=data.split(',')
    for d in data:
        if(is_english(d)==False):
            df.drop(index,inplace=True)
            break
```

```python
for index, data in df['TranslatedIngredients'].iteritems():
    ingredients = []
    # print(data)
    for i in data.split(','):
        t=numhyphnumb(i)
        t=remove_numbers(t)
        t=remove_after_hyphen(t)
        t.strip()
        ingredients.append(t)
    ingredients = ','.join(ingredients)
    df['TranslatedIngredients'][index]=ingredients
# df['TranslatedIngredients'].to_csv('ingredients_no_digit_lasthyphen.csv')
```

```python
stop=set()
stop.add('tsps')
stop.add('tbsps')
stop.add('tsp')
stop.add('tbsp')
stop.add('cups')
stop.add('cup')
stop.add('tablespoons')
stop.add('teaspoons')
stop.add('teaspoon')
stop.add('tablespoon')
stop.add('pinches')
stop.add('pinch')
stop.add('pounds')
stop.add('pound')
stop.add('grams')
stop.add('kgs')
stop.add('kg')
stop.add(' g')
stop.add('g ')
stop.add('kilo')
stop.add('kilos')
stop.add('liter')
stop.add('liters')
stop.add('litre')
stop.add('litres')
stop.add('ml ')
stop.add(' ml')

for index,data in df['TranslatedIngredients'].iteritems():
    data=data.split(',')
    ingredients=[]
    for d in data:
        d=d.lower()
        d=d.strip()
        for i in stop:
            if i in d:
                d=d.replace(i,'')
        d=d.strip()
        ingredients.append(d)
    ingredients = ','.join(ingredients)
    df['TranslatedIngredients'][index]=ingredients
# df['TranslatedIngredients'].to_csv('ingredients_measurements.csv')
```

```python
for index,data in df['TranslatedIngredients'].iteritems():
    data=data.split(',')
    ingredients=[]
    for d in data:
        d=d.lower()
        if d.startswith('s '):
            d=d[2:]
        d=d.strip()
        ingredients.append(d)
    ingredients = ','.join(ingredients)
    df['TranslatedIngredients'][index]=ingredients
# df['TranslatedIngredients'].to_csv('ingredients_measurements1.csv')
```

```python
for index,data in df['TranslatedIngredients'].iteritems():
    data=data.split(',')
    ingredients=[]
    for d in data:
        d=d.lower()
        if '(split)' in d:
            d=d.replace('(split)','')
        if 'split' in d:
            d=d.replace('split','')
        d=d.strip()
        ingredients.append(d)
    ingredients = ','.join(ingredients)
    df['TranslatedIngredients'][index]=ingredients
# df['TranslatedIngredients'].to_csv('ingredients_split.csv')
```

```python
from nltk import word_tokenize
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))
stop_words.add('per')
stop_words.add('taste')
for index,data in df['TranslatedIngredients'].iteritems():
    data=data.split(',')
    ingredients=[]
    for d in data:
        d=d.lower()
        d=d.strip()
        if ' ( ' in d:
            d=d.replace(' ( ','/')
        if '(' in d:
            d=d.replace('(','/')
        if ' (' in d:
            d=d.replace(' (','/')
        if '( ' in d:
            d=d.replace('( ','/')
        if ')' in d:
            d=d.replace(')','')
        d=d.strip()
        d=word_tokenize(d)
        d=[w for w in d if not w in stop_words]
        d=' '.join(d)
        ingredients.append(d)
    ingredients = ','.join(ingredients)
    df['TranslatedIngredients'][index]=ingredients
# df['TranslatedIngredients'].to_csv('ingredients_stopwords.csv')
```

```python
ingredient_alias=[]
dont_want=['raw','fresh','dried','dry','whole','unsalted','uncooked','unpealed','sun','diced','chopped','flattened','ground','peeled','sliced','sliced','minced','crushed','grated','shredded','squeezed','juiced','zested','pitted','seeded','halved','quartered','cubed','mashed','melted','softened','chilled','frozen','thawed','cooked','boiled','steamed','roasted','baked','toasted','fried','grilled','broiled','blanche','make','little']
for index,data in df['TranslatedIngredients'].iteritems():
    data=data.split(',')
    for d in data:
        if '/' in d:
            alias=[]
            d=d.split('/')
            if '' in d:
                d.remove('')
            if ' ' in d:
                d.remove(' ')
            if '+' in d:
                d.remove('+')
            if '-' in d:
                d.remove('-')
            if ' ' in d:
                d.remove('')
            d=[w for w in d if not w in dont_want]
            if len(d)>1:
                for i in d:
                    i=i.strip()
                    alias.append(i)
                ingredient_alias.append(alias)
# ingredient_alias
```

```python
# function remove punctuations from a string
import string
def remove_punctuation(s):
    s = ''.join([i for i in s if i not in frozenset(string.punctuation)])
    return s
# remove_punctuation('salt')
```

```python
ingred_count={}
c=0
for index,data in df['TranslatedIngredients'].iteritems():
    data=data.split(',')
    print(data)
    for d in data:
        if d=='salt':
            c+=1
        if '/' in d:
            d=d.split('/')
            if '' in d:
                d.remove('')
            if ' ' in d:
                d.remove(' ')
            if '+' in d:
                d.remove('+')
            if '-' in d:
                d.remove('-')
            if ' ' in d:
                d.remove(' ')
            d=[w for w in d if not w in dont_want]
            i=d[0].strip()
            if remove_punctuation(i).strip() in ingred_count:
                ingred_count[remove_punctuation(i).strip()]+=1
            else:
                ingred_count[remove_punctuation(i).strip()]=1
        else:
            d=d.strip()
            if remove_punctuation(d).strip() in ingred_count:
                ingred_count[remove_punctuation(d).strip()]+=1
            else:
                ingred_count[remove_punctuation(d).strip()]=1
# ingred_count
```

```python
sorted_ingred_count = {k: v for k, v in sorted(ingred_count.items(), key=lambda item: item[1], reverse=True)}
len(sorted_ingred_count)
```

```python
import copy
ingred_count_copy=copy.deepcopy(ingred_count)
```

```python
for i in ingredient_alias:
    for j in range(1,len(i)):
        if i[0] in ingred_count and i[j] in ingred_count:
            ingred_count_copy[i[0]]+=ingred_count_copy[i[j]]
            ingred_count_copy[i[j]]=0
            ingred_count.pop(i[j])
ingred_count=copy.deepcopy(ingred_count_copy)
for i in ingred_count_copy:
    if ingred_count_copy[i]==0:
        ingred_count.pop(i)
# ingred_count
```

```python
# sort the dictionary by value
sorted_ingred_count = {k: v for k, v in sorted(ingred_count.items(), key=lambda item: item[1], reverse=True)}
len(sorted_ingred_count)
```

```python
# stem all the keys in sorted_ingred_count
# copy the dictionary
sorted_copy=copy.deepcopy(sorted_ingred_count)
from nltk.stem import PorterStemmer
ps = PorterStemmer()
for i in sorted_ingred_count:
    if i!=ps.stem(i) and ps.stem(i) in sorted_copy:
        sorted_copy[ps.stem(i)]+=sorted_copy[i]
        sorted_copy.pop(i)
    elif i==ps.stem(i) and ps.stem(i) in sorted_copy:
        continue
    elif ps.stem(i) not in sorted_copy:
        sorted_copy[ps.stem(i)]=sorted_copy[i]
        sorted_copy.pop(i)
# sorted_copy
```

```python
# sort the dictionary by value
# sorted_ingred_c = {}
sorted_ingred_c = {k: v for k, v in sorted(sorted_copy.items(), key=lambda item: item[1], reverse=True)}
# sorted_ingred_c
```

```python
# plot graph using 150 values of sorted dictionary 
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
plt.bar(list(sorted_ingred_c.keys())[0:150], list(sorted_ingred_c.values())[0:150])
plt.xticks(rotation=90)
plt.show()
```

```python
# set threshold 
thresh=887
# keys with count higher than threshold put in a new list
ingred_count_thresh = {}
for key in sorted_ingred_c.keys():
    if sorted_ingred_c[key] > thresh:
        ingred_count_thresh[key] = sorted_ingred_c[key]
common_ingred=list(ingred_count_thresh.keys())
ingred_count_thresh
```

```python
for index,data in df['TranslatedIngredients'].iteritems():
    data=data.split(',')
    dc=copy.deepcopy(data)
    ex=[]
    # print(data)
    for da in data:
        # print("hi")
        d=copy.deepcopy(da)
        # print(da)
        flag=0
        if '/' in d:
            d=d.split('/')
            if '' in d:
                d.remove('')
            if ' ' in d:
                d.remove(' ')
            if '+' in d:
                d.remove('+')
            if '-' in d:
                d.remove('-')
            if ' ' in d:
                d.remove(' ')
            d=[w for w in d if not w in dont_want]
            for i in range(len(d)):
                d[i]=remove_punctuation(d[i]).strip()
            for i in d:
                i=i.strip()
                if ps.stem(i) in common_ingred:
                    dc.remove(da)
                    flag=1
                    break
            if flag!=1:
                d=','.join(d)
                dc.remove(da)
                ex.append(d)
        else:
            d=d.strip()
            d=ps.stem(d)
            if remove_punctuation(d).strip() in common_ingred:
                dc.remove(da)
    dc.extend(ex)
    data=','.join(dc)
    df['TranslatedIngredients'][index]=data
    # print(data)
```

```python
import nltk

def combine_proper_nouns(sentence):
    tagged_words = nltk.pos_tag(nltk.word_tokenize(sentence))
    proper_nouns = []
    combined_proper_nouns = []

    for word, tag in tagged_words:
        # print(word,tag)
        if (word not in dontstartwith) and (tag == 'NNP' or tag == 'NNPS') :  # Check if the word is a proper noun
            proper_nouns.append(word)
        else:
            if proper_nouns:
                combined_proper_nouns.append(' '.join(proper_nouns))
                proper_nouns = []

    if proper_nouns:  # Handle proper nouns at the end of the sentence
        combined_proper_nouns.append(' '.join(proper_nouns))

    combined_proper_nouns= [x for x in combined_proper_nouns if x not in dontwantalone]
    return combined_proper_nouns

# Example usage
# sentence = "John Smith is from New York City and works at Google."
# result = combine_proper_nouns(sentence)
# print(result)
```

```python
import nltk

def combine_nouns(sentence):
    tagged_words = nltk.pos_tag(nltk.word_tokenize(sentence))
    nouns = []
    combined_nouns = []

    for word, tag in tagged_words:
        if word not in dontstartwith and tag.startswith('NN'):  # Check if the word is a noun
            nouns.append(word)
        else:
            if nouns:
                combined_nouns.append(' '.join(nouns))
                nouns = []

    if nouns:  # Handle nouns at the end of the sentence
        combined_nouns.append(' '.join(nouns))

    combined_nouns= [x for x in combined_nouns if x not in dontwantalone]
    return combined_nouns

# Example usage
# sentence = "The cat is chasing the mouse in the garden."
# result = combine_nouns(sentence)
# print(result)
```

```python
# function to compare similarity of two strings
from difflib import SequenceMatcher
def similar(a, b):
    return SequenceMatcher(None, a, b).ratio()
# similar('paneer','paneer butter masala')
```

```python
##version2 , output = output_combinations1.csv
def get_dishnames(text):
    text = text.replace('.', " ")
    dish_names = []
    # Tokenize the text into words
    tokens = nltk.word_tokenize(text)

    # Perform POS tagging
    pos_tags = nltk.pos_tag(tokens)

    # Print the tagged words
    i = 0
    n = len(pos_tags)
    while i<n:
        if pos_tags[i][1] in ['NNP', 'NNPS']:
            if i!=n-1:
                dish = ""
                while i<n and pos_tags[i][1] in ['NNP', 'NNPS']:
                    if pos_tags[i][0] not in ['Serve', 'Mix', 'Add']:
                        dish+=pos_tags[i][0]
                        dish+=" "
                        i+=1
                    else:
                        break

                    
                    
            else:
                dish = pos_tags[i][0]
            dish_names.append(dish.strip())
        i+=1
    dish_names = [x for x in dish_names if x!='']
    return dish_names
```

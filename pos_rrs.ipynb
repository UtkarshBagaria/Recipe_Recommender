```python
##version2 , output = output_combinations1.csv
def get_dishnames(text):
    text = text.replace('.', " ")
    dish_names = []
    # Tokenize the text into words
    tokens = nltk.word_tokenize(text)

    # Perform POS tagging
    pos_tags = nltk.pos_tag(tokens)

    # Print the tagged words
    i = 0
    n = len(pos_tags)
    while i<n:
        if pos_tags[i][1] in ['NNP', 'NNPS']:
            if i!=n-1:
                dish = ""
                while i<n and pos_tags[i][1] in ['NNP', 'NNPS']:
                    if pos_tags[i][0] not in ['Serve', 'Mix', 'Add']:
                        dish+=pos_tags[i][0]
                        dish+=" "
                        i+=1
                    else:
                        break

                    
                    
            else:
                dish = pos_tags[i][0]
            dish_names.append(dish.strip())
        i+=1
    dish_names = [x for x in dish_names if x!='']
    return dish_names
```

```python
##version4 , output = output_combinations4.csv
def get_dishnames2(text):
    text = text.replace('.', " ")
    text = text.replace('!', " ")
    dish_names = []
    # Tokenize the text into words
    tokens = nltk.word_tokenize(text)

    # Perform POS tagging
    pos_tags = nltk.pos_tag(tokens)

    # Print the tagged words
    i = 0
    n = len(pos_tags)
    while i<n:
        if pos_tags[i][1] in ['NNP', 'NNPS']:
            if i!=n-1:
                dish = ""
                while i<n and pos_tags[i][1] in ['NNP', 'NNPS']:
                    if pos_tags[i][0] not in ['Serve', 'Mix', 'Add', 'Sunday', 'South', 'North']:
                        dish+=pos_tags[i][0]
                        dish+=" "
                        i+=1
                    else:
                        break

                    
                    
            else:
                dish = pos_tags[i][0]
            dish_names.append(dish.strip())
        i+=1
    dish_names = [x for x in dish_names if x!='']


    if len(dish_names)==0:
        j = 0
        while j<n:
            if pos_tags[j][1] in ['NN', 'NNS']:
                if j!=n-1:
                    dish = ""
                    while j<n and pos_tags[j][1] in ['NN', 'NNS']:
                        dish+=pos_tags[j][0]
                        dish+=" "
                        j+=1

                else:
                    dish = pos_tags[j][0]
                dish_names.append(dish.strip())
            j+=1

        dish_names = [x for x in dish_names if x not in ['dinner', 'cup', 'Enjoy','office lunchbox', 'Box', '']]


    return dish_names
```

```python
#version5
def get_dishnames3(text):
    text = text.replace('.', " ")
    text = text.replace('!', " ")
    dish_names = []
    # Tokenize the text into words
    tokens = nltk.word_tokenize(text)

    # Perform POS tagging
    pos_tags = nltk.pos_tag(tokens)

    # Print the tagged words
    i = 0
    n = len(pos_tags)
    while i<n:
        if pos_tags[i][1] in ['NNP', 'NNPS']:
            if i!=n-1:
                dish = ""
                while i<n and pos_tags[i][1] in ['NNP', 'NNPS', 'CC', 'JJ']:
                    if pos_tags[i][0] not in ['Serve', 'Mix', 'Add', 'Sunday']:
                        dish+=pos_tags[i][0]
                        dish+=" "
                        i+=1
                    else:
                        break

                    
                    
            else:
                dish = pos_tags[i][0]
            dish_names.append(dish.strip())
        i+=1
    dish_names = [x for x in dish_names if x!='']
    dish_names = [x for x in dish_names if x not in ['serve', 'box', 'meals', 'South', 'North', '']]


    if len(dish_names)==0:
        j = 0
        while j<n:
            if pos_tags[j][1] in ['NN', 'NNS']:
                if j!=n-1:
                    dish = ""
                    while j<n and pos_tags[j][1] in ['NN', 'NNS']:
                        dish+=pos_tags[j][0]
                        dish+=" "
                        j+=1

                else:
                    dish = pos_tags[j][0]
                dish_names.append(dish.strip())
            j+=1

        dish_names = [x for x in dish_names if x not in ['dinner', 'cup', 'Enjoy',]]


    return dish_names
```

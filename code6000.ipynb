```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag

def remove_adjectives_verbs_adverbs_prepositions(sentence):
    # Tokenize the sentence into words
    words = word_tokenize(sentence)
    
    # Perform part-of-speech tagging
    tagged_words = pos_tag(words)
    
    # Filter out adjectives, verbs, adverbs, and prepositions
    filtered_words = [word for word, tag in tagged_words if tag not in ['JJ', 'JJR', 'JJS', 'VB', 'VBD', 'VBG', 'VBN', 'VBP', 'VBZ', 'RB', 'RBR', 'RBS', 'IN']]
    
    # Join the filtered words back into a sentence
    filtered_sentence = ' '.join(filtered_words)
    
    return filtered_sentence
```

```python
import re

def filter_ingredients(recipe):
    ingredients = []
    # keep only the ingedient names
    for ingredient in recipe:
        # make everything lowercase
        ingredient = ingredient.lower()
        ingredient = ingredient.split(' ')
        ingredient = [word for word in ingredient if not word.isdigit()]
        ingredient = [word for word in ingredient if not word=='teaspoon']
        ingredient = [word for word in ingredient if not word=='teaspoons']
        ingredient = [word for word in ingredient if not word=='tablespoon']
        ingredient = [word for word in ingredient if not word=='tablespoons']
        ingredient = [word for word in ingredient if not word=='cup']
        ingredient = [word for word in ingredient if not word=='cups']
        ingredient = [word for word in ingredient if not word=='pound']
        ingredient = [word for word in ingredient if not word=='pounds']
        ingredient = [word for word in ingredient if not word=='gram']
        ingredient = [word for word in ingredient if not word=='grams']
        ingredient = [word for word in ingredient if not word=='g']
        ingredient = [word for word in ingredient if not word=='kg']
        ingredient = [word for word in ingredient if not word=='kgs']
        ingredient = [word for word in ingredient if not word=='ml']
        ingredient = [word for word in ingredient if not word=='ltr']
        ingredient = [word for word in ingredient if not word=='ltrs']
        ingredient = [word for word in ingredient if not word=='liter']
        ingredient = [word for word in ingredient if not word=='liters']
        ingredient = [word for word in ingredient if not word=='l']
        ingredient = [word for word in ingredient if not word=='tbsp']
        ingredient = [word for word in ingredient if not word=='tbsps']
        ingredient = [word for word in ingredient if not word=='tsp']
        ingredient = [word for word in ingredient if not word=='tsps']
        ingredient = ' '.join(ingredient)
        if 'to taste' in ingredient:
            ingredient = ingredient.replace('to taste','')
        if 'as per taste' in ingredient:
            ingredient = ingredient.replace('as per taste','')
        # Remove punctuation marks and digits
        ingredient = re.sub(r'[^\w\s]+|\d+', '', ingredient)
        # Remove extra whitespaces between words
        ingredient = re.sub(r'\s+', ' ', ingredient).strip()
        # Remove adjectives, verbs, adverbs, and prepositions
        ingredient = remove_adjectives_verbs_adverbs_prepositions(ingredient)
        ingredients.append(ingredient)

            
    return ingredients
# print(df1['TranslatedIngredients'][1])
# print(filter_ingredients(df1['TranslatedIngredients'][1]))
for index, data in df1['TranslatedIngredients'].iteritems():
    data=filter_ingredients(data)
    # print(data)
    df1.at[index, 'TranslatedIngredients'] = data
    # print(df1['TranslatedIngredients'][index])
# print(df1['TranslatedIngredients'][1])
# print(filter_ingredients(df1['TranslatedIngredients'][1]))
```

```python
# eda on ingredients
ingred_count = {}
for index, data in df1['TranslatedIngredients'].iteritems():
    for d in data:
        if d in ingred_count:
            ingred_count[d] += 1
        else:
            ingred_count[d] = 1
```

```python
# sort in descending order according to the count
ingred_count = {k: v for k, v in sorted(ingred_count.items(), key=lambda item: item[1], reverse=True)}
ingred_count.pop('')
ingred_count
```

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
stop_words = set(stopwords.words('english'))
stop_words.add('taste')
stop_words.add('tastes')
stop_words.add('a')
stop_words.add('an')
stop_words.add('the')
stop_words.add('as')
stop_words.add('per')
stop_words.add('to')
stop_words.add('pinch')
stop_words.add('pinches')
stop_words.add('cup')
ps = PorterStemmer()
def cleaned_key(sen):
    sen = sen.split(' ')
    for i in sen:
        if i in stop_words:
            sen.remove(i)
    sen = [ps.stem(word) for word in sen]
    sen = ' '.join(sen)
    # print(sen)
    return sen
# print(stop_words)
```

```python
# remove stopwords and stem the keys in the dictionary to check if they are same 
# if they are same then merge the keys by summing their values
import copy
ingred_count_lower=copy.deepcopy(ingred_count)
# make all the keys in ingred_count lowercase
for x in ingred_count:
    ingred_count_lower[x.lower()] = ingred_count_lower.pop(x)
ingred_count_copy = copy.deepcopy(ingred_count_lower)

for key in ingred_count:
    key=key.lower()
    key1=remove_adjectives_verbs_adverbs_prepositions(key)
    new_key = cleaned_key(key1)
    if new_key in ingred_count_copy and new_key!=key:
        ingred_count_copy[new_key] += ingred_count_copy[key]
        ingred_count_copy.pop(key)
    elif new_key!=key:
        ingred_count_copy[new_key] = ingred_count_copy[key]
        ingred_count_copy.pop(key)
# sort in descending order according to the count
ingred_count_copy = {k: v for k, v in sorted(ingred_count_copy.items(), key=lambda item: item[1], reverse=True)}
ingred_count_copy.pop('')
ingred_count_copy
    
```

```python
# set threshold 
thresh=846
# keys with count higher than threshold put in a new list
ingred_count_thresh = {}
for key in ingred_count_copy:
    if ingred_count_copy[key] > thresh:
        ingred_count_thresh[key] = ingred_count_copy[key]
common_ingred=list(ingred_count_thresh.keys())
common_ingred
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag

def remove_adjectives_verbs_adverbs_prepositions(sentence):
    # Tokenize the sentence into words
    words = word_tokenize(sentence)
    
    # Perform part-of-speech tagging
    tagged_words = pos_tag(words)
    
    # Filter out adjectives, verbs, adverbs, and prepositions
    filtered_words = [word for word, tag in tagged_words if tag not in ['JJ', 'JJR', 'JJS', 'VB', 'VBD', 'VBG', 'VBN', 'VBP', 'VBZ', 'RB', 'RBR', 'RBS', 'IN']]
    
    # Join the filtered words back into a sentence
    filtered_sentence = ' '.join(filtered_words)
    
    return filtered_sentence
```

```python
import enchant
dictionary = enchant.Dict("en_US")
dictionary.add('dal')
dictionary.add('roti')
dictionary.add('rasam')
dictionary.add('dosa')
dictionary.add('idli')
dictionary.add('chutney')
dictionary.check('biryani')
def check_english_words(sentence):
    words = sentence.split()
    non_english_words = []
    
    for word in words:
        if not dictionary.check(word):
            non_english_words.append(word)
    
    return non_english_words
```

```python
def keep_after_hyphen(sentence):
    parts = sentence.split('-')
    if(len(parts) == 1):
        parts = sentence.split('(')
    # remove punctuation marks from the two parts
    parts = [re.sub(r'[^\w\s]+', '', part) for part in parts]
    if len(parts) > 1:
        if(check_english_words(parts[0]) == []):
            return parts[0]
        else:
            return parts[1]        
    else:
        return sentence
```

```python
def cleanRecipe(recipe):
    # make everything lowercase
    recipe = recipe.lower()
    recipe = keep_after_hyphen(recipe)
    # Remove punctuation marks and digits
    recipe = re.sub(r'[^\w\s]+|\d+', '', recipe)
    # Remove extra whitespaces between words
    recipe = re.sub(r'\s+', ' ', recipe).strip()
    # remove the occurrence of the word recipe
    recipe = recipe.replace('asfoitida','asafoetida ')
    recipe = recipe.replace('recipe', '')
    recipe = recipe.replace('recipes', '')
    recipe = recipe.replace('breakfast', '')
    recipe = recipe.replace('lunch', '')
    recipe = recipe.replace('dinner', '')
    recipe = recipe.replace('snack', '')
    recipe = recipe.replace('snacks', '')
    recipe = recipe.replace('dessert', '')
    recipe = recipe.replace('desserts', '')
    recipe = recipe.replace('appetizer', '')
    recipe = recipe.replace('appetizers', '')
    recipe = recipe.replace('side dish', '')
    recipe = recipe.replace('side dishes', '')
    recipe = recipe.replace('main dish', '')
    recipe = recipe.replace('main dishes', '')
    recipe = recipe.replace('style', '')
    recipe = recipe.replace('hindi', '')
    # Remove adjectives, verbs, adverbs, and prepositions
    recipe = remove_adjectives_verbs_adverbs_prepositions(recipe)
    return recipe
# cleanRecipe(df1['TranslatedRecipeName'][13])
```

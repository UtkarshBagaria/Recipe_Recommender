```python
import re

def is_english(text):
    english_pattern = re.compile(r'^[A-Za-z0-9.,!?;:(){}\[\]<>@#$%^&*_+=\-|\\/~`"\'£€¥₹₽\s]*$')
    return bool(re.match(english_pattern, text))
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag

def remove_adjectives_verbs_adverbs_prepositions(sentence):
    # Tokenize the sentence into words
    words = word_tokenize(sentence)
    
    # Perform part-of-speech tagging
    tagged_words = pos_tag(words)
    
    # Filter out adjectives, verbs, adverbs, and prepositions
    filtered_words = [word for word, tag in tagged_words if tag not in ['JJ', 'JJR', 'JJS', 'VB', 'VBD', 'VBG', 'VBN', 'VBP', 'VBZ', 'RB', 'RBR', 'RBS', 'IN']]
    
    # Join the filtered words back into a sentence
    filtered_sentence = ' '.join(filtered_words)
    
    return filtered_sentence
```

```python
import re

def filter_ingredients(recipe):
    ingredients = []
    # keep only the ingedient names
    for ingredient in recipe:
        # make everything lowercase
        ingredient = ingredient.lower()
        ingredient = ingredient.split(' ')
        ingredient = [word for word in ingredient if not word.isdigit()]
        ingredient = [word for word in ingredient if not word=='teaspoon']
        ingredient = [word for word in ingredient if not word=='teaspoons']
        ingredient = [word for word in ingredient if not word=='tablespoon']
        ingredient = [word for word in ingredient if not word=='tablespoons']
        ingredient = [word for word in ingredient if not word=='cup']
        ingredient = [word for word in ingredient if not word=='cups']
        ingredient = [word for word in ingredient if not word=='pound']
        ingredient = [word for word in ingredient if not word=='pounds']
        ingredient = [word for word in ingredient if not word=='gram']
        ingredient = [word for word in ingredient if not word=='grams']
        ingredient = [word for word in ingredient if not word=='g']
        ingredient = [word for word in ingredient if not word=='kg']
        ingredient = [word for word in ingredient if not word=='kgs']
        ingredient = [word for word in ingredient if not word=='ml']
        ingredient = [word for word in ingredient if not word=='ltr']
        ingredient = [word for word in ingredient if not word=='ltrs']
        ingredient = [word for word in ingredient if not word=='liter']
        ingredient = [word for word in ingredient if not word=='liters']
        ingredient = [word for word in ingredient if not word=='l']
        ingredient = [word for word in ingredient if not word=='tbsp']
        ingredient = [word for word in ingredient if not word=='tbsps']
        ingredient = [word for word in ingredient if not word=='tsp']
        ingredient = [word for word in ingredient if not word=='tsps']
        ingredient = ' '.join(ingredient)
        if 'to taste' in ingredient:
            ingredient = ingredient.replace('to taste','')
        if 'as per taste' in ingredient:
            ingredient = ingredient.replace('as per taste','')
        # Remove punctuation marks and digits
        ingredient = re.sub(r'[^\w\s]+|\d+', '', ingredient)
        # Remove extra whitespaces between words
        ingredient = re.sub(r'\s+', ' ', ingredient).strip()
        # Remove adjectives, verbs, adverbs, and prepositions
        ingredient = remove_adjectives_verbs_adverbs_prepositions(ingredient)
        ingredients.append(ingredient)

            
    return ingredients
# print(df1['TranslatedIngredients'][1])
# print(filter_ingredients(df1['TranslatedIngredients'][1]))
for index, data in df1['TranslatedIngredients'].iteritems():
    data=filter_ingredients(data)
    # print(data)
    df1.at[index, 'TranslatedIngredients'] = data
    print(df1['TranslatedIngredients'][index])
# print(df1['TranslatedIngredients'][1])
# print(filter_ingredients(df1['TranslatedIngredients'][1]))
```

```python
for index, data in df1['TranslatedIngredients'].iteritems():
    # Modify the value
    data = data.strip()

    # Remove newlines and tabs
    data = data.replace('\n', '').replace('\t', '')

    # Remove extra spaces
    data = ' '.join(data.split())

    # Split the data by commas
    ingredients = data.split(',')

    # Remove leading and trailing whitespaces from each ingredient
    ingredients = [ingredient.strip() for ingredient in ingredients]

    df1.at[index, 'TranslatedIngredients'] = ingredients
df1['TranslatedIngredients'].head()
```

```python
import re

def is_english(text):
    english_pattern = re.compile(r'^[A-Za-z0-9.,!?;:(){}\[\]<>@#$%^&*_+=\-|\\/~`"\'£€¥₹₽\s]*$')
    return bool(re.match(english_pattern, text))
```

```python
for index, data in df1['TranslatedIngredients'].iteritems():
    for d in data:
        if(is_english(d)==False):
            df1.drop(index,inplace=True)
            break
```

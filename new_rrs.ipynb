```python
import enchant
dictionary = enchant.Dict("en_US")
dictionary.add('dal')
dictionary.add('roti')
def check_english_words(sentence):
    words = sentence.split()
    non_english_words = []
    
    for word in words:
        if not dictionary.check(word):
            non_english_words.append(word)
    
    return non_english_words
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
import re

def remove_adjectives_verbs_adverbs_prepositions(sentence):
    # Tokenize the sentence into words
    words = word_tokenize(sentence)
    
    # Perform part-of-speech tagging
    tagged_words = pos_tag(words)
    
    # Filter out adjectives, verbs, adverbs, and prepositions
    filtered_words = [word for word, tag in tagged_words if tag not in ['JJ', 'JJR', 'JJS', 'VB', 'VBD', 'VBG', 'VBN', 'VBP', 'VBZ', 'RB', 'RBR', 'RBS', 'IN']]
    
    # Join the filtered words back into a sentence
    filtered_sentence = ' '.join(filtered_words)
    
    return filtered_sentence
```

```python
def keep_after_hyphen(sentence):
    parts = sentence.split('-')
    if(len(parts) == 1):
        parts = sentence.split('(')
    # remove punctuation marks from the two parts
    parts = [re.sub(r'[^\w\s]+', '', part) for part in parts]
    if len(parts) > 1:
        if(check_english_words(parts[0]) == []):
            return parts[0]
        else:
            return parts[1]        
    else:
        return sentence
```

```python
def cleanRecipe(recipe):
    # make everything lowercase
    recipe = recipe.lower()
    recipe = keep_after_hyphen(recipe)
    # Remove punctuation marks and digits
    recipe = re.sub(r'[^\w\s]+|\d+', '', recipe)
    # Remove extra whitespaces between words
    recipe = re.sub(r'\s+', ' ', recipe).strip()
    # remove the occurrence of the word recipe
    recipe = recipe.replace('asfoitida','asafoetida ')
    recipe = recipe.replace('recipe', '')
    recipe = recipe.replace('recipes', '')
    recipe = recipe.replace('breakfast', '')
    recipe = recipe.replace('lunch', '')
    recipe = recipe.replace('dinner', '')
    recipe = recipe.replace('snack', '')
    recipe = recipe.replace('snacks', '')
    recipe = recipe.replace('dessert', '')
    recipe = recipe.replace('desserts', '')
    recipe = recipe.replace('appetizer', '')
    recipe = recipe.replace('appetizers', '')
    recipe = recipe.replace('side dish', '')
    recipe = recipe.replace('side dishes', '')
    recipe = recipe.replace('main dish', '')
    recipe = recipe.replace('main dishes', '')
    recipe = recipe.replace('style', '')
    recipe = recipe.replace('hindi', '')
    # Remove adjectives, verbs, adverbs, and prepositions
    recipe = remove_adjectives_verbs_adverbs_prepositions(recipe)
    return recipe
# cleanRecipe(df1['TranslatedRecipeName'][13])
```

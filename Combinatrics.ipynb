```python
import pandas as pd
df1=pd.read_csv('Working_file11.csv')
df1.head()
```

# Combinations Code

```python
import re
for index,data in df1['TranslatedInstructions'].iteritems():
    data= data.strip()
    # Remove extra whitespaces between words
    data = re.sub(r'\s+', ' ', data)
    # Find the index of the second last period or exclamation mark
    # Find all occurrences of periods and exclamation marks
    c=0
    for i in range(len(data)-1,-1,-1):
        if(data[i]=='!' or data[i]=='.'):
            c+=1
        if(c==2):
            print((data[i+1:]).strip())
            break
```
